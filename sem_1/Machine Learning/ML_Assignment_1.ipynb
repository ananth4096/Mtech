```python
#Split your data into train(80%) and test(20%)
def tts(x, y, test_size=0.2, random_state=None):
    np.random.seed(random_state)# Set the random seed for reproducibility
    num_samples = x.shape[0]# Get the total number of samples in the dataset
    num_test_samples = int(test_size * num_samples)# Calculate the number of samples to be used for the test set
    shuffled_indices = np.random.permutation(num_samples)# Generate shuffled indices to randomly split the data
    train_indices, test_indices = shuffled_indices[num_test_samples:], shuffled_indices[:num_test_samples]# Generate shuffled indices to randomly split the data
    x_train, x_test = x.iloc[train_indices], x.iloc[test_indices] #Split the input features 'X' into train and test sets based on the indices
    y_train, y_test = y.iloc[train_indices], y.iloc[test_indices] #Split the target variable 'Y' into train and test sets based on the indices
    return x_train, x_test, y_train, y_test
```

```python

class KNN_Algorithm:
    def fit(self, X, y):
        self.x_train = X # Store the training data
        self.y_train = y # Store the corresponding labels
    
    def hamming_distance(self, s1, s2): # Calculate the Hamming distance between two strings s1 and s2.
        return sum(c1 != c2 for c1, c2 in zip(s1, s2)) # Hamming distance counts the positions at which the characters differ.
    
    def predict(self, X, k=3):
        y_pred = [] # List to store the predicted labels for each input
        for x in X:
            # Calculate Hamming distances between the input 'x' and all training data points
            distances = [self.hamming_distance(x, x_train) for x_train in self.x_train]
             # Find the indices of the 'k' nearest neighbors based on calculated Hamming distances
            k_indices = np.argsort(distances)[:k]  # Select the 'k' smallest distances
            # Get labels of 'k' nearest neighbors
            k_nearest_labels = self.y_train[k_indices]
            # Predict the label based on the most common label among neighbors
            # Use np.bincount to count occurrences of each label and choose the most common one
            y_pred.append(np.bincount(k_nearest_labels).argmax())# Add the predicted label to the list
        return np.array(y_pred)# Return the predicted labels as an array
```
