```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# Lists to store R2 scores for each cluster
r2_scores = [0] * optimal_k  # Initialize with zeros
U_values = []
S_values = []
VT_values = []
```

```python
# Tensor Factorization (SVD) and Similarity Calculation for each cluster
for cluster_id in range(optimal_k):
    cluster_mask = (cluster_labels == cluster_id)
    cluster_ratings = ratings[cluster_mask]

    # Calculate the mean (centroid) of the cluster
    cluster_centroid = np.mean(cluster_ratings, axis=0)

    # Subtract the cluster centroid from the cluster ratings
    centered_ratings = cluster_ratings - cluster_centroid

    # Store the centered ratings for later use
    if cluster_id == 0:
        all_centered_ratings = centered_ratings
    else:
        all_centered_ratings = np.concatenate((all_centered_ratings, centered_ratings), axis=0)
```

```python
# Calculate SVD for all centered ratings using NumPy's built-in function
U, S, VT = np.linalg.svd(all_centered_ratings, full_matrices=False)
print(f"Cluster {cluster_id} - U: {U.shape}, S: {S.shape}, VT: {VT.shape}")
```

```python
# Store the U, S, VT values for each cluster
start_index = 0
for cluster_id in range(optimal_k):
    cluster_mask = (cluster_labels == cluster_id)
    num_ratings = np.sum(cluster_mask)

    U_values.append(U[start_index:start_index+num_ratings])
    S_values.append(S)
    VT_values.append(VT)

    start_index += num_ratings
```

```python
# Preference Prediction (CART) and R-squared (R2) Calculation for each cluster
for cluster_id in range(optimal_k):
    cluster_mask = (cluster_labels == cluster_id)
    cluster_ratings = ratings[cluster_mask]

    # Retrieve the U, S, VT values for the current cluster
    U = U_values[cluster_id]
    VT = VT_values[cluster_id]

    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(cluster_ratings, overall_ratings[cluster_mask], test_size=0.2, random_state=42)

    # Transform the ratings using the U and VT matrices
    centered_ratings = X_train - np.mean(X_train, axis=0)
    transformed_ratings = np.dot(np.dot(centered_ratings, VT.T), np.linalg.inv(np.diag(S)))

    # Train a Decision Tree Regressor
    regressor = DecisionTreeRegressor()
    regressor.fit(transformed_ratings, y_train)

    # Transform the test ratings using the U and VT matrices
    centered_test_ratings = X_test - np.mean(X_test, axis=0)
    transformed_test_ratings = np.dot(np.dot(centered_test_ratings, VT.T), np.linalg.inv(np.diag(S)))

    # Predict the ratings for the test set
    y_pred = regressor.predict(transformed_test_ratings)

    # Calculate R-squared for regression evaluation
    r2 = r2_score(y_test, y_pred)
    r2_scores[cluster_id] = r2

    print(f"Cluster {cluster_id} - R-squared (R2) Score: {r2}")
```

```python
# Plot the R-squared scores
plt.figure(figsize=(8, 6))
plt.plot(range(optimal_k), r2_scores, marker='o')
plt.title('R-squared (R2) Score for Each Cluster')
plt.xlabel('Cluster ID')
plt.ylabel('R2 Score')
plt.grid(True)
plt.show()
```
