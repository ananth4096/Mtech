## Take any data set, find dinensionality reduction using SVD

```python
import numpy as np

# Create a synthetic dataset with 100 samples and 3 features
np.random.seed(0)  # For reproducibility
X = np.random.rand(100, 3)
X
```

```python
# Perform SVD on the dataset
U, S, Vt = np.linalg.svd(X, full_matrices=False)

# Choose the number of components (dimensions) to keep
n_components = 2

# Reduce dimensionality by selecting the top components
X_reduced = np.dot(U[:, :n_components], np.diag(S[:n_components])).dot(Vt[:n_components, :])
```

```python
X_reduced.shape
```

```python
import matplotlib.pyplot as plt

# Original dataset
plt.scatter(X[:, 0], X[:, 1], label='Original Data')

# Reduced dataset
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], label='Reduced Data', marker='x')

plt.legend()
plt.title('Original vs. Reduced Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
